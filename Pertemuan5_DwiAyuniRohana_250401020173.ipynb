```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# --- LANGKAH 1: Mengatur Tema Dasar Visualisasi ---
sns.set_theme(style="whitegrid")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 10

# --- LANGKAH 2: Memuat Dataset ---
data_tips = sns.load_dataset('tips')

# --- LANGKAH 3: Inisialisasi Arsitektur Dashboard-
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))

# Memberikan Judul Utama Dashboard
fig.suptitle('Dashboard Analisis Transaksi & Tip Restoran (Tips Dataset)',
             fontsize=18, fontweight='bold', y=0.96, color='#2c3e50')

# ------------------------------------------------------------------------------
# GRAFIK 1 (Kiri Atas): Bar Plot - Rata-rata Total Tagihan per Hari
# ------------------------------------------------------------------------------
avg_bill_day = data_tips.groupby('day', observed=False)['total_bill'].mean().reset_index()
sns.barplot(data=avg_bill_day, x='day', y='total_bill', hue='day', palette='Blues_d', legend=False, ax=axes[0, 0])
axes[0, 0].set_title('1. Rata-rata Total Tagihan Per Hari', fontsize=12, fontweight='bold', pad=10)
axes[0, 0].set_xlabel('Hari')
axes[0, 0].set_ylabel('Rata-rata Tagihan ($)')

# ------------------------------------------------------------------------------
# GRAFIK 2 (Kanan Atas): Histogram & KDE - Distribusi Frekuensi Total Tagihan
# ------------------------------------------------------------------------------
sns.histplot(data=data_tips, x='total_bill', kde=True, color='#1abc9c', ax=axes[0, 1])
axes[0, 1].set_title('2. Distribusi Frekuensi Total Tagihan Pelanggan', fontsize=12, fontweight='bold', pad=10)
axes[0, 1].set_xlabel('Total Tagihan ($)')
axes[0, 1].set_ylabel('Frekuensi Kunjungan')

# ------------------------------------------------------------------------------
# GRAFIK 3 (Kiri Bawah): Boxplot - Rentang Pengeluaran Berdasarkan Jenis Kelamin & Hari
sns.boxplot(data=data_tips, x='day', y='total_bill', hue='sex', palette='Set2', ax=axes[1, 0])
axes[1, 0].set_title('3. Rentang Pengeluaran Pria vs Wanita per Hari', fontsize=12, fontweight='bold', pad=10)
axes[1, 0].set_xlabel('Hari')
axes[1, 0].set_ylabel('Total Tagihan ($)')
axes[1, 0].legend(title='Jenis Kelamin')

# ------------------------------------------------------------------------------
# GRAFIK 4 (Kanan Bawah): Scatter Plot - Hubungan Total Tagihan vs Tip yang Diberikan
# ------------------------------------------------------------------------------
sns.scatterplot(data=data_tips, x='total_bill', y='tip', hue='size', palette='viridis', alpha=0.8, ax=axes[1, 1])
axes[1, 1].set_title('4. Korelasi Total Tagihan vs Besaran Tip Pelanggan', fontsize=12, fontweight='bold', pad=10)
axes[1, 1].set_xlabel('Total Tagihan ($)')
axes[1, 1].set_ylabel('Tip ($)')
axes[1, 1].legend(title='Ukuran Rombongan (Size)')
plt.tight_layout(rect=[0, 0.03, 1, 0.93])

# Menampilkan Dashboard secara Utuh
plt.show()
```

1. What? (Apa yang ditampilkan?)
Dashboard ini menampilkan empat analisis visual dari data transaksi restoran (dataset 'tips'). Grafik ini mencakup rata-rata total tagihan per hari, distribusi frekuensi tagihan (yang didominasi kisaran $10–$20), perbandingan rentang pengeluaran antara pria dan wanita per harinya, serta korelasi positif antara total tagihan dan besaran tip.

2. So What? (Apa maknanya?)
Dari grafik tersebut, terlihat jelas bahwa puncak transaksi dan pengeluaran tertinggi terjadi pada akhir pekan (Sabtu dan Minggu). Selain itu, pelanggan pria sering kali mencatat tagihan yang sangat tinggi (outliers), dan terbukti bahwa semakin besar total pesanan meja beserta jumlah rombongannya, semakin besar pula tip yang mereka tinggalkan.

3. Now What? (Apa langkah selanjutnya?)
Berdasarkan pola data tersebut, manajemen restoran dapat mengambil tindakan bisnis seperti: menambah jadwal staf pelayan pada hari Sabtu dan Minggu untuk mengimbangi lonjakan pengunjung, serta membuat promo "Paket Keluarga" untuk menarik rombongan besar karena terbukti menghasilkan tagihan dan tip yang paling maksimal.
