```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df.head()
```

```python
"Pertanyaan 1 : Berapa Jumlah baris dan kolom dataset titanic"
"Jawaban : diketahui ada 891 baris dan 12 kolom"

print(df.shape)
```

```python
"Pertanyaan 2 : Kolom mana yang memiliki nilai kosong"
"Jawaban : diketahui bahwa kolom yang ada nilai kosong adalah Age, Cabin dan Embarked"

print(df.isnull().sum())
```

```python
"Pertanyaan 3 : Berapa persen penumpang yang selamat"
"Jawaban : diketahui 38% yang selamat"

persen_selamat = df['Survived'].mean() * 100
print(persen_selamat,"%")
```

```python
"Pertanyaan 4 : Rata-rata usian penumpang wanita kelas 1 yang selamat"
"Jawaban : diperoleh rata-rata usia wanita kelas 1 yang selamat adalah 34 tahun"

wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]
print(wk1['Age'].mean())
```

```python
"Pertanyaan 5: Kelas peumpang mana yang tingkat keselamatan tertinggi"
"Jawaban : diperoleh bahwa kelas 1 mempunyai tingkat keselamatan yang paling tinggi 0.63 atau 63%"

hasil = df.groupby('Pclass').agg(
total=('PassengerId','count'),
selamat=('Survived','mean'),
).round(3)
print(hasil)
```

```python
"Pertanyaan 6: Menampilkan hasil dari groupby('Pclass')['Survived'].mean()"
hasil = df.groupby('Pclass')['Survived'].mean()
print(hasil)
```
